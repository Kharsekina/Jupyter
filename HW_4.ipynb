```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('kc-house-data.csv')
```

Изучите стоимости недвижимости.

```python
plt.figure(figsize=(8,6))
plt.hist(df['price'], bins=100)
plt.title('Стоимость недвижимости')
plt.xlabel('Стоимость, млн')
plt.ylabel('Количество');
```

Изучите распределение квадратуры жилой.

```python

plt.figure(figsize=(8,6))
sns.histplot(df['sqft_living'], bins=50)
plt.title('Распределение жилплощади')
plt.xlabel('Жилплощадь')
plt.ylabel('Количество');
```

Изучите распределение года постройки.

```python
bild_year = df['yr_built'].value_counts().reset_index().sort_values(by='yr_built')
bild_year.head()
```

```python

plt.figure(figsize=(8,6))
plt.plot(bild_year['yr_built'],bild_year['count'])
plt.title('Год постройки')
plt.xlabel('Год')
plt.ylabel('Количество');
```

 Изучите распределение домов от наличия вида на набережную. (Меньше процента домов имеют вид на набережную)

```python
data_water = df['waterfront'].value_counts()
data_water.index = ['no', 'yes']
plt.figure(figsize=(6,4))

plt.pie(data_water,autopct='%1.10f%%')

plt.legend(data_water.index);
```

 Изучите распределение этажей домов.

```python
data_floors = df['floors'].value_counts().reset_index().sort_values(by='floors')
data_floors.head()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=data_floors['floors'], y=data_floors['count'])

plt.title('Этажность')
plt.xlabel('Этажи')
plt.ylabel('Количество');
```

Изучите распределение состояния домов.

```python
data_cond = df['condition'].value_counts()
data_cond.index = ['1', '2', '3', '4', '5']
plt.figure(figsize=(6,4))

plt.pie(data_cond,autopct='%1.10f%%')

plt.legend(data_cond.index);
```
